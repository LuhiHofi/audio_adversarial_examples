#### Spektrogramy

```python
import whisper

model_small = whisper.load_model("small")
PGD_001_path = "/home/luhi/lukas-hofman/data/transcriptions/PGD-01/small/"

nat = whisper.load_audio(PGD_001_path + "nat.wav")
adv = whisper.load_audio(PGD_001_path + "adv.wav")

pertubation = adv - nat

import librosa
import numpy as np

# Compute perturbation
perturbation = adv - nat

# Parameters
sample_rate = 16000
hop_length = 512
n_fft = 400
n_mels = 80

# Generate Mel spectrogram
mel_spectrogram = librosa.feature.melspectrogram(
    y=perturbation,
    sr=sample_rate,
    n_fft=n_fft,
    hop_length=hop_length,
    n_mels=n_mels
)

# Convert to dB scale (log-Mel)
log_mel_spectrogram = librosa.power_to_db(mel_spectrogram, ref=np.max)

# Plot
plt.figure(figsize=(10, 4))
librosa.display.specshow(
    log_mel_spectrogram,
    sr=sample_rate,
    hop_length=hop_length,
    x_axis="time",
    y_axis="mel"
)
plt.colorbar(format="%+2.0f dB")
plt.title("Log-Mel Spektrogram perturbace PGD útoku")
plt.tight_layout()
plt.show()
```
